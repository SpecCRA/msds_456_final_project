# Setup

```python
import pandas as pd
import numpy as np
```

```python
from nba_api.stats.static import players
from nba_api.stats.static import teams
```

```python
# get list of active players
active_players = players.get_active_players()

# get all teams
all_teams = teams.get_teams()

# get blazers
blazers = teams.find_teams_by_city('Portland')

# set current season
curr_season = '2018-19'
```

```python
blazers[0]

# get team id
blazers_id = blazers[0]['id']
```

# Stats to gather

## Team

2. Usage per player
3. Pace relative to other teams

## Players

1. point % of team totals

```python
# Usage per player per 100 poss
from nba_api.stats.endpoints import LeagueDashPlayerStats
player_adv_stats = LeagueDashPlayerStats(measure_type_detailed_defense='Advanced', 
                                            per_mode_detailed='Per100Possessions',
                                            team_id_nullable=blazers_id,
                                            season=curr_season).get_data_frames()[0]

player_adv_stats.head()
```

```python
list(player_adv_stats)
```

```python
blazers_adv_stats = LeagueDashPlayerStats(measure_type_detailed_defense='Usage',
                                    season=curr_season,
                                    team_id_nullable=blazers_id).get_data_frames()[0]
blazers_adv_stats
```
